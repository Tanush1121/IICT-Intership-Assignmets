```python
# ==============================================================================
# ENVIRONMENT SETUP: GENERATING THE TITANIC DATASET FILE
# ==============================================================================
import pandas as pd
import numpy as np

# 1. Generate local data so our assignment scripts execute cleanly
np.random.seed(42)
mock_data = pd.DataFrame({
    'PassengerId': range(1, 101),
    'Survived': np.random.choice([0, 1], size=100, p=[0.6, 0.4]),
    'Pclass': np.random.choice([1, 2, 3], size=100, p=[0.2, 0.3, 0.5]),
    'Sex': np.random.choice(['male', 'female'], size=100),
    'Age': np.random.normal(30, 12, size=100).clip(1, 80),
    'SibSp': np.random.choice([0, 1, 2], size=100, p=[0.7, 0.2, 0.1]),
    'Fare': np.random.exponential(32, size=100).clip(7, 500)
})

# Introduce missing values to match the preprocessing assignment challenges
mock_data.loc[np.random.choice(100, 10), 'Age'] = np.nan

# 2. Save it as 'titanic.csv' into the Colab cloud environment
mock_data.to_csv("titanic.csv", index=False)
print("✅ Environment Ready: 'titanic.csv' has been generated in the cloud backend!")
```
