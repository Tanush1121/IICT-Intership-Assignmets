```python
# SETUP CELL: Re-create the dataset for Notebook 2
import pandas as pd
import numpy as np

np.random.seed(42)
mock_data = pd.DataFrame({
    'PassengerId': range(1, 101),
    'Survived': np.random.choice([0, 1], size=100, p=[0.6, 0.4]),
    'Pclass': np.random.choice([1, 2, 3], size=100, p=[0.2, 0.3, 0.5]),
    'Sex': np.random.choice(['male', 'female'], size=100),
    'Age': np.random.normal(30, 12, size=100).clip(1, 80),
    'SibSp': np.random.choice([0, 1, 2], size=100, p=[0.7, 0.2, 0.1]),
    'Fare': np.random.exponential(32, size=100).clip(7, 500)
})
mock_data.to_csv("titanic.csv", index=False)
print("✅ Dataset file ready for visualization!")
```

```python
# ==============================================================================
# ASSIGNMENT 2: DATA VISUALIZATION LAB (DAY 3)
# ==============================================================================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the data file
df = pd.read_csv("titanic.csv")

# 2. Build the Exploratory Chart layout
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x='Age',
    y='Fare',
    hue='Survived',
    palette='coolwarm',
    alpha=0.85,
    s=80,       # Adjusts dot size for legibility
    edgecolor='w'
)

# 3. Add Context and Structural Labels
plt.title("Exploratory Analysis: Passenger Age vs Ticket Fare Profile", fontsize=14, pad=15)
plt.xlabel("Passenger Age (Years)", fontsize=12)
plt.ylabel("Ticket Fare (USD)", fontsize=12)
plt.legend(title="Survival Status", labels=["Perished (0)", "Survived (1)"])
plt.grid(True, linestyle='--', alpha=0.5)

# 4. Display the chart beautifully on screen
plt.show()
print("==================================================================")
print("ASSIGNMENT 2 COMPLETED SUCCESSFULLY!")
print("==================================================================")
```
