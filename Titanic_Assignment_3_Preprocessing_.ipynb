```python
# ==============================================================================
# ASSIGNMENT 3: DATA PREPROCESSING & CATEGORICAL MAPPING (DAY 5)
# ==============================================================================
import pandas as pd
import numpy as np

print("==================================================================")
print("RUNNING ASSIGNMENT 3: DATA PREPROCESSING PIPELINE")
print("==================================================================\n")

# --- PART A: GENERATING THE RAW DATA EXTRACTION ---
np.random.seed(42)
raw_data = pd.DataFrame({
    'PassengerId': range(1, 101),
    'Survived': np.random.choice([0, 1], size=100, p=[0.6, 0.4]),
    'Pclass': np.random.choice([1, 2, 3], size=100, p=[0.2, 0.3, 0.5]),
    'Sex': np.random.choice(['male', 'female'], size=100),
    'Age': np.random.normal(30, 12, size=100).clip(1, 80),
    'SibSp': np.random.choice([0, 1, 2], size=100, p=[0.7, 0.2, 0.1]),
    'Fare': np.random.exponential(32, size=100).clip(7, 500)
})
# Inject missing values into Age to simulate real-world data corruption
raw_data.loc[np.random.choice(100, 15), 'Age'] = np.nan
raw_data.to_csv("titanic_raw.csv", index=False)


# --- PART B: THE PREPROCESSING PIPELINE ---
# 1. Load the raw dataset
df = pd.read_csv("titanic_raw.csv")

print("🔍 [BEFORE] Missing Values Count:")
print(df[['Age', 'Sex', 'Pclass']].isnull().sum())

# 2. Impute Missing Values (Replacing missing Age values with the Column Mean)
age_mean = df['Age'].mean()
df['Age'] = df['Age'].fillna(age_mean)
print(f"\n Strategy: Filled missing 'Age' rows with the average age: {age_mean:.2f} years.")

# 3. Handle Categorical Text Data (Mapping text strings into binary 0 and 1)
# Machine Learning models cannot process the word strings "male" and "female" directly
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
print("Strategy: Encoded categorical 'Sex' column values into numerical bits (male = 0, female = 1).")

print("\n------------------------------------------------------------------")
print("🔍 [AFTER] Verification Statistics:")
print("------------------------------------------------------------------")
print("✅ Remaining Missing Values in Dataset:", df['Age'].isnull().sum())
print("\n📋 Cleaned Dataframe Preview (Notice the 'Sex' column integers):")
print(df[['PassengerId', 'Sex', 'Age', 'Pclass', 'Fare']].head())

print("\n==================================================================")
print("ASSIGNMENT 3 COMPLETED SUCCESSFULLY!")
print("==================================================================")
```
