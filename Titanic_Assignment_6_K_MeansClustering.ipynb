```python
# ==============================================================================
# ASSIGNMENT 6: UNSUPERVISED LEARNING — K-MEANS CLUSTERING (DAY 14)
# ==============================================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

print("==================================================================")
print("RUNNING ASSIGNMENT 6: K-MEANS CLUSTERING PIPELINE")
print("==================================================================\n")

# --- PART A: GENERATE ANALYTICAL REFERENCE DATA ---
np.random.seed(42)
df = pd.DataFrame({
    'Age': np.random.normal(30, 12, size=150).clip(1, 80),
    'Fare': np.random.exponential(35, size=150).clip(8, 250)
})

# --- PART B: FEATURE SCALING ---
# K-Means calculates distances, so we scale features to prevent 'Fare' from dominating 'Age'
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[['Age', 'Fare']])

# --- PART C: TRAINING THE K-MEANS ALGORITHM ---
# We set n_clusters=3 to segregate passengers into 3 distinct structural tiers
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X_scaled)
print("✅ Success: K-Means algorithm has completed passenger clustering calculations.")

# --- PART D: ANALYZING THE CLUSTER ARCHETYPES ---
print("\n------------------------------------------------------------------")
print("📊 PASSENGER CLUSTER CHARACTERISTIC PROFILES:")
print("------------------------------------------------------------------")
cluster_profile = df.groupby('Cluster').mean()
print(cluster_profile)

print("\n💡 Cluster Interpretation Guide:")
print("   - High Fare Cluster: Represents the wealthy elite passengers.")
print("   - Younger/Lower Fare Cluster: Represents youth/working-class travelers.")
print("   - Older/Moderate Fare Cluster: Represents established mid-tier travelers.")

# --- PART E: VISUALIZING THE CLUSTERS ---
plt.figure(figsize=(9, 5))
sns.scatterplot(
    data=df,
    x='Age',
    y='Fare',
    hue='Cluster',
    palette='Set1',
    s=70,
    alpha=0.8
)
plt.title("Unsupervised Learning: Passenger Segmentation via K-Means", fontsize=13)
plt.xlabel("Passenger Age")
plt.ylabel("Ticket Fare")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()

print("\n==================================================================")
print("ASSIGNMENT 6 COMPLETED SUCCESSFULLY!")
print("==================================================================")
```
